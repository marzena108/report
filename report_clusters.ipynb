<h2>Objectives:</h2>

- Embedding long chains with large gates (100 neurons) can provide reliable signal transmission, but it needs to be controlled (synaptic depression). With such sizes virtually no spontaneoulsy activity has been observed
- the above architecture assumes random connectivity (5%) with no control over reciprocal connections
- Literature shows that in cortical networks one can find a large amount of reciprocal connections (and triplets of such), much more than a theoretical random network would predict. (citation needed)
- pairs of reciprocally connected neurons tend to be sensitive to similar inputs (citation needed)
- reciprocal connections might also have stronger synaptic weights (citation needed)

Here, I will try to find out how the existence of reciprocal connections and triplets may change the dynamics of the network in the context of feedforward activity. A random network will be compared with a one with a fixed amount of reciprocal connections and triplets. Both will have a feedforward chain embedded and the fixed connections will be located within the chain.


<h2>Methods</h2>
<p>
Within a network of 2000 excitatory and 500 inhibitory neurons with 5% connectivity a short chain of 7 non-overlapping gates has been embedded. Each gate consists of 60 excitatory neurons and neurons in every gate (except for the first one) receive 40 randomly chosen connections from the preceding gate. Such a network comes in two variants. The first one is a basic network, with all the connections established randomly. In the second network, the connectivity along the chain is partly controlled. Within each gate a certain amount of triplets, mini nets as well as uni- and bidirectional connections have been introduced (sse the Figure 1 below). The weights of all the connections remained the same (a network with strenghtened reciprocal connections will come soon - at the moment the goal is to only investigate one parameter - the connectivity).
</p>
<p>
I created 60 instances of 'random network' and 60 instances of the clustered network. Each network's activity was recorded for 20s and the spike trains were used for the analysis.
</p>
<p>
Both networks had the parameters set to assure low firing rate (1.5Hz) and in both networks the signal along the chain is reliably propagated at the presence of a strong input to the first gate. The simulations were run to solely observe the spontaneous activity without any external inputs (apart form the usual external noise delivered to all the neurons in the network equally). 
</p>

<h3>A simple feedforward activity detection</h3>
<p>
The goal of this analysis was to find out whether the existence of clustered connectivity inside the gates would improve the overall feedforward activity at the absence of the input. The intuition was as follows: because the gates are explicitly connected (each neuron receives 40 connections from the preceding gate) it is expected that spikes coming from a gate, say number 3, will tend to excite the gate number 4. In the random network, neurons within a given gate are not connected to each other at all. Adding the clustered connections within a gate is expected to increase the amount of synchronous firing (or one after the other) within a given gate and thus sending a stronger signal to the next gate causing the gate to fire more than in the random, non-clustered network.

A simple test was run - For every spike that occurred along the chain at time t [ms] I count how many neurons occurred in a time window {t,t+10} and in which gate. The prediction is that if there is a spike in the gate 1, more spikes will occur in the same gate 1 as well as the following gate 2 than, say, gate 5. (same tests were done for different time windows too: {t+2,t+12},{t,t+5},{t+2,t+7}.

The plots below should be read as: A row 0 denotes a gate 0 and shows the total number of spikes that occured after spikes within the gate 0. With a chain embedded, it is expected that for a gate x a substantial number of spikes will occur in the gate x+1 due to the feedforward connections along the chain. With the clustered connections within gates, it is expected that spikes in x gate would cause more spikes in the same gate x. If the clustered connections indeed help to propagate the ff activity, the amount of total connection along the x,x+1 should be substantially larger that in the network without the clustered connections.


</p>

<h4>One more check</h4>

The result above suggests that no effect was seen after the insertion of the clusters and the feedforward activity was solely due to the ff connetions along the chain. To confirm this, one more network has been constructed. Again, 7 gates with clustered connections were embedded in a 2k network, but this time, the gates were not connected at all - no chain was established. Here the prediction was that if the clustered connections indeed change the dynamics, spikes within a given cluster should induce more spikes within this same cluster. 30 networks were created and the spikes during 20s of spontaneous activity were recorded. Again, the same simple algorithm was used to detect ff activity. Because the gates were not connected to each other, the only interesting point here is what happens along the diagonal. Instead of absolute numbers, I used the percentage - since we monitor the activity of 7 not-connected gates, each gate should be activated on average on 14% of trials. To be precise - when we scan for the spikes that occured after the spike in a gate 1, there is 14% probability that the following spike will come from gate 1, 14% that it will come from gate 2 and so on. Any deviation from the 14% should indicate that there might be some non-randomness involved.

The difference is really tiny, one can think that it is still a normal random network. Once we do a bit of maths things get clearer. What I actually did while embedding the clusters, I enforced 19 pairs of bidirectionally connected neurons per gate. 19x7=133 pairs or 133x2=266 connections. The total naumber of exc-exc connections in a 2k network is 2000x0.05x2000 = 200000. 266/200000 = 0.00133. Less than 1% of total connections were manipulated. From this perspective it should not be surprising that the overall network activity didn't change.

</p>
<p>
The question now is how to alter the network to reach a point where the effect might be visible. Increase the number of bidirectional connection within the gate, increase the size of the gate or organize the whole gate into a small-world mesh? Such an organization surely would enforce many triplets and single neurons would be involved in several reciprocal pairs. At the moment only 2 mini-nets are embedded and most of the bidirectional connections are isolated - coactivation of such pairs happens only coincidentally - there is no relation between them (apart from the ff connectivity along the chain).
