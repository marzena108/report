- Here, the analysis is extended to include 5k network and 10k network. 



<h2>Methods</h2>
<p>
Within a network of 5000 excitatory and 1000 inhibitory neurons with 5% connectivity a short chain of 7 (or 8) non-overlapping gates has been embedded. Each gate consists of 100 excitatory and 25 inhibitory neurons. Neurons in every gate (except for the first one) receive 60 randomly chosen connections from the preceding gate. Excitatory neurons receive 15 connections from randomly chosen inhibitory neurons from the same gate. Such a network comes in two variants. The first one is a basic network, with all the connections established randomly. In the second network, the connectivity along the chain is partly controlled. Within each gate I enforce a higher degree of connectivity - 40% - in such a way that every excitatory neuron receives input from 40% of neurons from the same gate. Reciprocal connetions and triplets are not explicitly controlled - instead the high ratio of connectivity imposes the existance of lots of bodirectionally connected pairs. 
</p>
<p>
Networks with 50% connectivity were briefly tested as well, but in such a case one adjustment had to be done. When sending a signal to the first gate, so far I would send it only to the excitatory population - it is just a matter of a few milisecond window whether the excitation is strong enough to cause sufficiently many neurons fire simultaneously to send a signal further or not. With the clustered connectivity, two things were observed: first of all, it turned out that with exactly the same network settings it takes significantly weaker input to activate the chain if the gates were clustered. Previously I would use a volley of 30-40 spikes per neuron, with clusters only 15-25 spikes are sufficient to fire a sequence. This is due to the fact that thanks to the clustered connectivity more neurons get excited over time and the excitation is retained within a gate for a short period of time. And that leads to the second observation - because the activity within the first gate is retained, after sending the signal further the gate remains active and keeps exciting the second gate - something that hasn't ben observed with random gates. To prevent from reactivating the chain, one has to add input also to the inhibitory population in the first gate which in turn will provide a delayed inhibition to the excitatory population in tthis gate. if the delay it too small, the gate fails to sufficiently excite the following gate, if the delay is too big - chain gets reactivated as long as the gate one remains active.

</p>
<p>
It seems that those features might be very useful - the input doesn't have to be very strong and synchronous to activate the first gate in a chain as the clustered connections inside the gate retain the activity sor some period of time. It is inhibition that can actually control the signal propagation - if it arrives quickly at the gate, it stops the neurons from sending the signals further, if it arrives a bit delayed - it allows for sending the signal and stops the neuron from continously firing and resending the signal. It should be noted though that such an addition of inhibitory control was needed only with the gates with 50% connectivity within the clusters. All the simulations described below had 40% connectivity at most and thus did not need suh a control - at the arrival of a strong input a gate would remain for some period of time but it was never long enough to cause the reactivation of the sequence. (But now I thnk it's quite interesting to see that happens with dense clusters so possibly more work should be done on clusters with >50% connectivity)
</p>

<h2>Another thought on inhibition</h2>

<p>While setting up the parameters of the 5k network to ensure low firing rates and successful propagation of a signal along the chain (to mimic the behaviour of the 2k network) one issue has been observed. It is essential to maintain the delayed inhibition along the chain as due to a rapidly increased activity within the whole network, the chain gets activated a number of times. 3 rasters below illustrate the case when a 8-gate chain was embedded in a 5k network. First case - gates consisted only of excitatory populations. Sending a strong input to the first gate activates the chain, but the gates along it get fired a couple of times due to elevated global activity. At some point all the neurons 'use up' and they all get deactivated.
Second raster shows a case when we include inhibitory populations to the gates (25 neurons each) and every excitatory neuron receives input from 5 inhibitory ones. Again, chains gets activated and gates fire a number of times. This time it all happens faster (25ms vs 40ms) but still the effect is not diserable.
Third case involves inhibition from 10 sources. Now the gates are activated as expected and there is no repeating. the signal transmission takes place once and no further activation is observed. 
</p>
<p>
From these tests it seems that in such a setup inhibition is essential to provide a balance for a suddenly elevated excitatory activity. In a 2k network such an effect was not observed at all - which proves that it makes sense to have large-sale simulations to detect various possible effects that might take place in large nets. Small toy examples might work fine, but they might not be robust enough to explain what happens in massive cortical networks. 5k networks that I am working on now should also be considered small - the hope is to learn something about the clustered connectivity and then move on to even bigger network to see whether the predictions still hold true.


</p>
<p>
Thanks to comparing 2k network with the 5k one, it's been decided that all the chains will contain both excitatory and inhibitory populations.
</p>

<h2>Actual simulations</h2>
<p>
So far it has been explained why the 40% connectivity within clusters has been used and why the delayed inhibition along the chain was reintroduced. With those features we could move on to test the predictions that thanks to the clustered connectivity one should observe more spontaneously generated sequential activity hopefully across long timescales. As before, for each case 60 networks were created and recorded for 20 seconds of activity (this choice was somewhat arbitrary - does it make sense to have many networks and average them over or have a few nets that can be run for minutes? (now as I write it I'm actually setting up a long simulation to record 2 and 4 minutes of activity to see how different the results can be). 
</p>

So I run 60 simulations for each case:
- a baseline: 5k net with 7 and then 8 gates (firstly i used 7 gates but then in a middle gate I could notice some weird symmetry so I decided to used 8 instead - see the Figure 6 below), random 5 percent connectivity across the whole network

- 5k net with 7 or 8 gates, clustered connectivity - 40% within each excitatory gate, inhibitory gates only from the second gate onwards (I think I did that out of laziness - my chain connecting algorithm was iterating gates staring from the second one and was using the 'previous gate' as source to connect to the 'current' gate. In that way no connectivity was established between the inhibitory and excitatory populations within the gate one. As explaind earlier, inhibition in the first gate was somewhat superfluous as I would always give an input only to the excitatory neurons, so the inh-exc connections could never really be harnessed in a chain propagation. So the omitting on the first gate was never recognied as a bug.

- 5k net with 7 or 8 gates, clustered connectivity - 40% within each excitatory gate, inhibitory gates in EVERY gate. Interestingly, the case described above raised some questions that could only be answered by running the simulations with the inhibitory gate added to the first gate as well. This issue will be discussed more thoroughly in subsequent sections.

- 5k net with 7 or 8 gates, clustered connectivity - 20%  or 30% within each excitatory gate - this is to see whether gradually increasing the level of density of connections would systematically increase te sequential activity detection.

- 5k net 7 or 8 gates, 40% clustered connectivity, no chain at all - this is to see whether the clusters themselves would induce some sort of ff activity or any considerable changes in the overall dynamics

- 5k net, 7 or 8 gates, 40% clustered connectivity, no FF chain, only inh-exc connections retained from the chain structure. This example was to check how much influence has the directed inhibition on its own.
